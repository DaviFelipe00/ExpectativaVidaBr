```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
df = pd.read_csv("C:\\Users\\davif\\Downloads\\idh.csv")

df_ano_idh = df[['ano_referencia', 'expectativa_de_vida']].rename(columns={'ano_referencia': 'Ano', 'expectativa_de_vida': 'Expectativa de vida'})

df_ano_idh.tail(10)
```

```python
# Criar o grfico
plt.figure(figsize=(10,5))
plt.plot(df["ano_referencia"], df["expectativa_de_vida"], marker="o", linestyle="-", color="g", label="Expectativa de vida brasileira")

# Adicionar rtulos e ttulo
plt.xlabel("Ano")
plt.ylabel("ndice de expectativa")
plt.title("Evoluo da expectativa de vida dos brasileiros")
plt.legend()
plt.grid(True)

# Mostrar o grfico
plt.show()
```
